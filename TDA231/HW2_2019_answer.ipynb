**Question 2**:

Some useful identities are listed as:

$$\frac{\partial (\mathbf{w}^T\mathbf{x})}{\partial \mathbf{w}}=\mathbf{x}$$

$$\frac{\partial (\mathbf{x}^T\mathbf{w})}{\partial \mathbf{w}}=\mathbf{x}$$

$$\frac{\partial(\mathbf{w}^T\mathbf{w})}{\partial \mathbf{w}}=2\mathbf{w}$$

$$\frac{\partial (\mathbf{w}^T\mathbf{C} \mathbf{w})}{\mathbf{w}}=2\mathbf{C}\mathbf{w}$$
 
The result of taking the partial derivative with
 respect to a vector is a vector where each element is the partial derivative with respect to one parameter:
 
 
$$\frac{\partial \mathcal{L}}{\partial \mathbf{w}}=\left[
\begin{array}{c}
           \frac{\partial \mathcal{L}}{\partial w_0}\\
           \frac{\partial \mathcal{L}}{\partial w_1}\\
           \vdots\\
           \frac{\partial \mathcal{L}}{\partial w_p}
\end{array}
\right]$$

$$\frac{\partial \mathcal{L}}{\partial \mathbf{w}}=\frac{\partial ((\mathbf{y}-\mathbf{X}\mathbf{w})^T(\mathbf{y}-\mathbf{X}\mathbf{w})+\alpha|\mathbf{w}|_2^2)}{\partial \mathbf{w}}$$
$$=\frac{\partial ((\mathbf{y}-\mathbf{X}\mathbf{w})^T(\mathbf{y}-\mathbf{X}\mathbf{w})}{\partial\mathbf{w}}+\frac{\partial (\alpha \mathbf{w}^T\mathbf{w})}{\partial \mathbf{w}}$$
$$=2\mathbf{X}^T\mathbf{X}\mathbf{w}-2\mathbf{X}^T\mathbf{y}+2\alpha \mathbf{w}$$

**Question 3**:

The optimization problem equals that:

$$\frac{\partial \mathcal{L}}{\partial \mathbf{w}}=0$$
$$2\mathbf{X}^T\mathbf{X}\mathbf{w}-2\mathbf{X}^T\mathbf{y}+2\alpha \mathbf{w}=0$$
$$\mathbf{X}^T\mathbf{X}\mathbf{w}+\alpha \mathbf{I}\mathbf{w}=\mathbf{X}^T\mathbf{y}$$
$$(\mathbf{X}^T\mathbf{X}+\alpha \mathbf{I})\mathbf{w}=\mathbf{X}^T\mathbf{y}$$
$$(\mathbf{X}^T\mathbf{X}+\alpha \mathbf{I})^{-1}(\mathbf{X}^T\mathbf{X}+\alpha \mathbf{I})\mathbf{w}=(\mathbf{X}^T\mathbf{X}+\alpha \mathbf{I})^{-1}\mathbf{X}^T\mathbf{y}$$
$$\mathbf{w^*}=(\mathbf{X}^T\mathbf{X}+\alpha \mathbf{I})^{-1}\mathbf{X}^T\mathbf{y}$$

Preknowledge:

For a matrix $A$ and $B$, assume matrix $A$ is a $m \times n$ matrix. $rank(AB) \leq min(rank(A),rank(B)) \leq rank(A) \leq min(m,n)$

We need to invert a matrix $\mathbf{X}^T\mathbf{X}+\alpha \mathbf{I}$ and to make sure that it is invertible,  $\alpha > 0$

When $\alpha=0$, i.e., in the case of ordinary least squares, the condition that $p > N$ causes the sample covariance matrix $\mathbf{X}^T\mathbf{X}$ to not have full rank and so it cannot be inverted to yield a unique solution. This is why there can be an infinitude of solutions to the ordinary least squares problem when $\mathbf{X}^T\mathbf{X}$. However, when $\alpha > 0$, i.e., when ridge regression is used, the addition of $\alpha$to the sample covariance matrix ensures that all of its eigenvalues will be strictly greater than 0. In other words, it becomes invertible, and the solution becomes unique.

```
# Make it possible to show plots in the notebooks.
%matplotlib inline
import numpy as np
from sklearn import datasets
import matplotlib
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

def fit_linear_with_regularization(X, y, alpha):
  
  # calculate w according to the math expression above
  matrix_size = np.shape(X)
  n = matrix_size[0]                # row count
  p = matrix_size[1]                # column count
  matrix_w0 = np.full((n, 1), 1)
  
  final_X = np.append(matrix_w0, X, axis=1)
  #print(np.shape(final_X))
  
  regulation = alpha * np.identity(p+1)
  #print(regulation)
  
  w = np.dot(np.linalg.pinv(np.dot(np.transpose(final_X), final_X) + regulation), np.dot(np.transpose(final_X), y))
  
  return w


def predict(X_test, w):
  
  # make prediction according to the math expression
  matrix_size = np.shape(X_test)
  n = matrix_size[0]               # row count
  p = matrix_size[1]               # column count
  matrix_w0 = np.full((n, 1), 1)
  
  final_X = np.append(matrix_w0, X_test, axis=1)
  
  x_predict = np.dot(final_X, w)
  
  return x_predict


def plot_prediction(X_test, y_test, y_pred):
  
  # Scatter plot the first feature of X_test (x-axis) and y_test (y-axis)
  X_test = np.array(X_test)
  x_test = X_test[:,0]

  line_up, = plt.plot(x_test, y_test, 'bo', label = 'Train')

  plt.title("Point observations")
  plt.xlabel("x")
  plt.ylabel("y")

  # Plot y_pred using the first feature of X_test as x-axis
  line_down, = plt.plot(x_test, y_pred, 'ro', label = 'Test')
  
  plt.legend([line_up, line_down],['Train','Test'])
  
  # Compute the mean squared error
  mse = 0.
  
  mse = mean_squared_error(y_test, y_pred)
  
  return mse
  

# Load the diabetes dataset
X, y = datasets.load_diabetes(return_X_y=True)

# normalize features
#from sklearn import preprocessing
#X = preprocessing.normalize(X)
#print(X)


# Split the dataset into training and test set
num_test_elements = 20

X_train = X[:-num_test_elements]
X_test = X[-num_test_elements:]

y_train = y[:-num_test_elements]
y_test = y[-num_test_elements:]


# Set alpha
alpha = 0.01

# Train using linear regression with regularization and find optimal model
w = fit_linear_with_regularization(X_train, y_train, alpha)

# Make predictions using the testing set X_test
y_pred = predict(X_test, w)

# Plots and mean squared error
mse = plot_prediction(X_test, y_test, y_pred)

print('Mean Squared error is ' + str(mse))

# Show the plot
plt.show();
```

Preknowledge:

Bayes Rule:
$$p(B\Big|A)=\frac{p(A\Big |B)\cdot p(B)}{p(A)}$$
$$p(A,B)=p(A)\cdot p(B \Big |A)$$
$$=p(B)\cdot p(A \Big | B)$$

posterior density can be expressed as,

$$p(\mathbf{w}\Big| \mathbf{X},\mathbf{y})=\frac{p(\mathbf{X},\mathbf{y}\Big| \mathbf{w})\cdot p(\mathbf{w})}{p(\mathbf{X},\mathbf{y})}$$
$$=\frac{p(\mathbf{X})\cdot p(\mathbf{y}\Big| \mathbf{w},\mathbf{X})\cdot p(\mathbf{w})}{p(\mathbf{X})\cdot p(\mathbf{y}\Big| \mathbf{X})}$$
$$=\frac{p(\mathbf{y}\Big| \mathbf{w},\mathbf{X})\cdot p(\mathbf{w})}{ p(\mathbf{y}\Big| \mathbf{X})}$$

Known conditions:

There is special relationship among $p(\mathbf{w})$, $p(\mathbf{y}\Big| \mathbf{w},\mathbf{X})$, $p(\mathbf{w}\Big| \mathbf{X},\mathbf{y})$ called conjugacy which means that prior is said to be conjugate to a likelihood and it results in a posterior of the same type of density as the prior.

In this case, prior is Gaussian, likelihood is Gaussian, so posterior must be Gaussian too.
 $p(\mathbf{w}) = \mathcal{N}(\mathbf{0}, \lambda^{-1} \mathbf{I})$ 
 
$p(\mathbf{y} | \mathbf{X}, \mathbf{w}) = \prod_{i=1}^n \mathcal{N}(\mathbf{w}^T \mathbf{x}_i, \gamma^{-1})$

We are trying to solve the log posterior:
$$lnp(\mathbf{w} \Big | \mathbf{y} ,\mathbf{X})=ln \mathcal{N}(\mu,\Sigma)$$
$$\propto ln exp(-\frac{1}{2}(\mathbf{w}-\mu)^T\Sigma^{-1}(\mathbf{w}-\mu))$$
$$=-\frac{1}{2}(\mathbf{w}-\mu)^T\Sigma^{-1}(\mathbf{w}-\mu)$$
$$=-\frac{1}{2}(\mathbf{w}^T\Sigma^{-1}\mathbf{w}-2\mathbf{w}^T\Sigma^{-1}\mu+\mu^T\Sigma^{-1}\mu)$$
