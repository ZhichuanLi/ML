## Bayes Rules
**Answer**:
According to the bayes rule:
$$ P(A \mid B) = \frac{P(B \mid A) \, P(A)}{P(B)} $$

We have below information from question:
$$P(B \mid A) = 0.992 $$
$$P(A) = 1/12000$$
$$P(A') =(12000-1)/12000 = 11999/12000.$$

So we can calcuate $P(A\mid B)$ as follows:
$$P(A\mid B) = \frac{P(B \mid A) \, P(A)}{P(B)} $$
$$=\frac{P(B \mid A) \, P(A)}{P(B \mid A) \, P(A) +P(B \mid A') \, P(A')} $$
$$=\frac{0.992*1/12000}{0.992*1/12000+0.008*11999/12000}$$
$$=\frac{0.992}{0.992+0.008*11999}$$
$$=0.992/96.984$$
$$=0.01$$

So the final result is roughly 0.01. i.e,  the probability you actually have the disease is 1%.

**Answer**:

We are given a dataset $X$, $X \sim \mbox{N}(\mu,2\sigma^2I)$, assume $2\sigma^2I=\Sigma$ here
$$X= \left[
  \begin{array}{ccc}
          X_1 &
          \cdots &
          X_p
  \end{array}
  \right]^T
$$
$$\mu=\left[
\begin{array}{c}
          \mu_1 \\
          \mu_2\\
          \vdots \\
          \mu_p
 \end{array}
 \right]$$
 $$\Sigma=\left[
 \begin{array}{ccc}
          2\sigma^2 & \cdots & 0\\
          \vdots & \ddots & \vdots\\
          0 & \cdots & 2\sigma^2
  \end{array}
\right]$$
$I$ is identity matrix, which means that components $X_n$ in $X$ is independent.
$$X_n \sim \mbox{N}(\mu_n,2\sigma^2)$$
So the maximum likelihood estimation is similar to normal univariate case whose MLE process is like below:
$$p(x|\mu)=\frac{1}{\sqrt{2\pi \sigma^2}}exp(-\frac{1}{2\sigma^2}(x-\mu)^2)$$
$$D=(x_1,x_2,x_3,...x_n)$$
$$p(D|\mu)=p(x_1,x_2,...x_n|\mu)=\prod_{i=1}^{n}p(x_i|\mu)$$
$$=(\frac{1}{\sqrt{2\pi \sigma^2}})^nexp(-\frac{1}{2\sigma^2}\sum_{i=1}^{n}(x_i-\mu)^2)$$
$$logL=log p(D|\mu)=-\frac{n}{2}log(2\pi\sigma^2)-\frac{1}{2\sigma^2}\sum_{i=1}^{n}(x_i-\mu)^2$$
$$0=\frac{\partial L}{\partial \mu}=\frac{1}{2\sigma^2}\sum_{i=1}^{n}2(x_i-\mu)=\frac{1}{\sigma^2}\sum_{i=1}^{n}(x_i-\mu)$$
$$\sum_{i=1}^{n}(x_i-\mu)=\sum_{i=1}^{n}x_i-n\mu=0$$
$$\hat{\mu}=\frac{1}{n}\sum_{i=1}^{n}x_i$$
$$0=\frac{\partial L}{\partial \sigma^2}=-\frac{n}{2}\frac{1}{2\pi\sigma^2}2\pi-\frac{1}{2}(-\frac{1}{(\sigma^2)^2}\sum_{i=1}^{n}(x_i-\mu)^2)$$
$$-\frac{n}{2}\frac{1}{\sigma^2}+\frac{1}{2}\frac{1}{(\sigma^2)^2}\sum_{i=1}^{n}(x_i-\mu)^2=0$$
$$\sigma^2=\frac{1}{n}\sum_{i=1}^{n}(x_i-\hat{\mu})^2$$

For our special multivariate Gaussian distribution case, we suppose $X$ is a $p*n$ matrix
$$\hat{\mu_i}=\frac{1}{n}\sum_{i=1}^{n}x_i $$
$$x_i\in X_i$$
$$\hat{\mu}=\left[
\begin{array}{c}
          \frac{1}{n}\sum X_1 \\
          \frac{1}{n}\sum X_2\\
          \vdots \\
          \frac{1}{n}\sum X_p
 \end{array}
 \right]=\frac{1}{n} \sum X_i$$
 
 $$N(\mu,\Sigma)=\frac{1}{(2\pi)^{p/2}| \Sigma |^{1/2}}exp(-\frac{1}{2}(X_i-\mu)^T\Sigma^{-1}(X_i-\mu))$$
 
$$log L=-\frac{N}{2}log[(2\pi)^p|\Sigma|]-\sum_{i=1}^{n}\frac{1}{2}(X_i-\mu)^T\Sigma^{-1}(X_i-\mu)$$

$$=-\frac{N}{2}log[(2\pi)^p(2\sigma^2)^p]-\frac{1}{2}\sum_{i=1}{n}\frac{1}{2\sigma^2}||X_i-\mu||^2$$
$$\frac{\partial}{\partial (2\sigma^2)}=-\frac{N}{2}\frac{p}{2\sigma^2}+\frac{1}{2}\frac{1}{(2\sigma^2)^2}\sum_{i=1}^{n}||X_i-\mu||^2=0$$
$$\hat{2\sigma^2}=\frac{\sum_{i=1}^{n}||X_i-\mu||^2}{np}$$


```python
# Answer to (b)
import numpy as np # numpy


def sge(X):
    """
    This function computes  the mean mu and standard deviation sigma 
    for a dataset coming from spherical Gaussian distribution N(mu, 2 sigma^2 I)

    @param X : A numpy array of size n x p where each row represents a p-dimensional data point
    e.g. X = [ [2, 1], [3, 7], [4 ,5] ] is a dataset with 3 samples having two coordinates each.
    @return mu: A numpy array of size p representing the estimated mean of the dataset
    @return sigma: A real number indicating the estimated standard deviation of the dataset
    """

    matrix_size = np.shape(X)
    row_count = matrix_size[0]           # row length of matrix X
    column_count = matrix_size[1]        # column length of matrix X
    mu = np.zeros(column_count)          # initiate mu as a numpy array of size p with value 0
    sum_mu = np.zeros(column_count)      # initiate sum of column as a numpy array of size p with value 0
    sigma = 0

    # calculate mu
    for m in X:
      for i in range(column_count):
        sum_mu[i] += m[i]                # get the sum of each column

    for j in range(len(sum_mu)):
      mu[j] = sum_mu[j] / row_count      # get the mean of each column
        
    #calculate sigma
    sum = 0

    for x in X:
      var_array = np.zeros(column_count)
      for k in range(column_count):
        var_array[k] = x[k] - mu[k]          # calculate Xi - mu

      res_dot = np.dot(var_array,var_array)  # doc product
      sum += res_dot                         # sum all the values

    sigma = np.sqrt(sum/(2 * row_count * column_count))  # 2 * sigma **2 = sum/(n * p)    
    
    return mu, sigma
```
