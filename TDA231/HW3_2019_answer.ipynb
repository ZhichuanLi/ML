### Your answer here:

Because all instances are projected onto a vector space of two real-valued features, i.e. all instances could be plotted on a X-Y plane, so the distance between two instances is just the distance between two plotted points.

**a)** if K=1

the test instance would be assigned to the class which its nearest neighbour belongs to, i.e. if the nearest point of the test point is positive(+), test instance will be assigned to positive class, vice versa.

   $\quad$if K=3

the test instance would be assigned to the class which more instances, wich means larger or equal to 2, among the '3 neighbours' belong to, i.e. if 2 of the 3 neighbours belong to positive(+), test instance will be assigned to positive class, if 3 of the 3 neighbours belong to positive(+), test instance will be assigned to positive class, vice versa.

   $\quad$if K=5

the test instance would be assigned to the class which more instances, wich means larger or equal to 3, among the '5 neighbours' belong to, i.e. if 3 of the 5 neighbours belong to positive(+), test instance will be assigned to positive class, if 4 of the 5 neighbours belong to positive(+), test instance will be assigned to positive class, if 5 of the 5 neighbours belong to positive(+), test instance will be assigned to positive class, vice versa.



**b)** the maximum value of K in this case should be $2\times min\{Num(+), Num(-)\}-1$, where $Num(+)$ represents the number of posotive instances in training set and $Num(-)$ represents the number of negative instances in training set.

Once $K > 2\times min\{Num(+), Num(-)\}$, assuming the number of positive instances is smaller, the maximum instance number that  the positive class could contribute is $Num(+)$, so the classification will always end up in 'negative class'. 

Furthermore, $K=2\times min\{Num(+), Num(-)\}$ should also be avoided to avoid a draw.

**Hints**:
- You compute $sigmoid( w^T x + b) = \frac{1}{1 + e^{-(w^T x + b)}}$ .
- You compute activation $A = \sigma(w^T X + b) = (a^{(0)}, a^{(1)}, ..., a^{(m-1)}, a^{(m)})$.
- You calculate the cost function: $J = -\frac{1}{m}\sum_{i=1}^{m}y^{(i)}\log(a^{(i)})+(1-y^{(i)})\log(1-a^{(i)})$.

Here are the two formulas you will be using: 

$$ \frac{\partial J}{\partial w} = \frac{1}{m}X(A-Y)^T$$
$$ \frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^m (a^{(i)}-y^{(i)})$$

- You write down the optimization function. The goal is to learn $w$ and $b$ by minimizing the cost function $J$. For a parameter $\theta$, the update rule is $ \theta = \theta - \alpha \text{ } d\theta$, where $\alpha$ is the learning rate.
- In prediction, you calculate $\hat{Y} = A = \sigma(w^T X + b)$.
- You may use np.exp, np.log(), np.dot(), etc. 

```
## Modeling and predicting

# GRADED FUNCTION: sigmoid
def sigmoid(z):
    """
    Return: s -- sigmoid(z)
    """
    s = 1 / (1 + np.exp(-z))
    
    return s
  
# GRADED FUNCTION: initialize_with_zeros
def initialize_with_zeros(dim):
    """
    Returns:
    w -- initialized vector of shape (dim, 1)
    b -- initialized to 0
    """    
    w = np.zeros((dim, 1))
    b = 0.
   
    assert(w.shape == (dim,1))
    assert(isinstance(b, float) or isinstance(b, int))
    
    return w, b
  
# GRADED FUNCTION: grad_cost
def grad_cost(w, b, X, Y):
  """
    Arguments:
    X -- data of size (number of features, number of examples)
    Y -- true "label" vector

    Return:
    cost -- negative log-likelihood cost for logistic regression
    dw -- gradient of the loss with respect to w, thus same shape as w
    db -- gradient of the loss with respect to b, thus same shape as b
  """
  m = X.shape[1]
  
  A = sigmoid(np.dot(np.transpose(w), X) + b)                                  
  cost = -(1 / m) * np.sum(Y * np.log(A) + (1 - Y) * np.log(1 - A))
  dw = (1 / m) * np.dot(X, np.transpose(A - Y))
  db = (1 / m) * np.sum((A - Y).reshape(1, m))                 
  
  assert(dw.shape == w.shape)
  assert(db.dtype == float)
  cost = np.squeeze(cost)
  assert(cost.shape == ())
    
  grads = {"dw": dw,
           "db": db}
    
  return grads, cost

 # GRADED FUNCTION: optimize
def optimize(w, b, X, Y, num_iterations, learning_rate, print_cost = False):
    """
    This function optimizes w and b by running a gradient descent algorithm
    """
    costs = []
    
    for i in range(num_iterations):
        
        grads, cost = grad_cost(w,b,X,Y)
        dw = grads["dw"]
        db = grads["db"]
        
        # update rule
        w = w - learning_rate * dw
        b = b - learning_rate * db
        
        if i % 100 == 0:
            costs.append(cost)     
        if print_cost and i % 100 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))
    
    params = {"w": w,
              "b": b}   
    grads = {"dw": dw,
             "db": db}
    
    return params, grads, costs
  
# GRADED FUNCTION: predict
def predict(w, b, X):
    '''
    Returns:
    Y_prediction -- a numpy array (vector) containing all predictions (0/1) for the examples in X
    '''    
    m = X.shape[1]
    Y_prediction = np.zeros((1,m))
    w = w.reshape(X.shape[0], 1)
    
    # Compute vector "A" predicting the probabilities of a cat being present in the picture
    A = sigmoid(np.dot(np.transpose(w), X) + b)
    
    for i in range(A.shape[1]):
        
        if A[0,i]<=0.5:
            Y_prediction[0,i]=0;
        if A[0,i]>0.5:
            Y_prediction[0,i]=1
    
    assert(Y_prediction.shape == (1, m))
    
    return Y_prediction
  
# GRADED FUNCTION: model
def model(X_train, Y_train, X_test, Y_test, num_iterations = 2000, learning_rate = 0.5, print_cost = False):
    """
    Builds the logistic regression model by calling the function you've implemented previously
    Returns: d -- dictionary containing information about the model.
    """    
    # initialize parameters with zeros
    p = X_train.shape[0]
    w, b = initialize_with_zeros(p)

    # Gradient descent
    parameters, grads, costs = optimize(w, b, X_train, Y_train, num_iterations, learning_rate, print_cost)
    
    w = parameters["w"]
    b = parameters["b"]
    
    Y_prediction_test = predict(w,b,X_test)
    Y_prediction_train = predict(w,b,X_train)

    print("train accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_train - Y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_test - Y_test)) * 100))
   
    d = {"costs": costs,
         "Y_prediction_test": Y_prediction_test, 
         "Y_prediction_train" : Y_prediction_train, 
         "w" : w, 
         "b" : b,
         "learning_rate" : learning_rate,
         "num_iterations": num_iterations}
    
    return d
```
